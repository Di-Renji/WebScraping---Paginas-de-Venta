```python
html_code = driver.page_source
```

```python
# Importamos y Usamos la libreria  BeautifulSoup 
from bs4 import BeautifulSoup
```

```python
# asignamos una lista vacia por cada columna que vamos a necesitar para nuestro dataframe
trabajos_cargo = []
trabajos_compania = []
trabajos_lugar = []
trabajos_descripcion = []
trabajos_hacecuanto = []

# Con una estructura For hacemos un recorrido de cada etiqueta con la respectiva clase del HTML que contiene el texto con la informacion para la columna que deseamos
pg_amount = 12
page =BeautifulSoup(driver.page_source, 'html.parser')
for i in range(0, pg_amount):
  for trabajos in page.findAll('article', class_='job clicky'):
      cargo = trabajos.find('h2', attrs={'href': ''})
      if cargo:
        trabajos_cargo.append(cargo.text)
      else:  trabajos_cargo.append('')

      compania = trabajos.find('p', attrs={'class': 'company'})
      if compania:
        trabajos_compania.append(compania.text)
      else:  trabajos_compania.append('')

      lugar= trabajos.find('ul', attrs={'class': 'location'})
      if lugar:
        trabajos_lugar.append(lugar.text)
      else:  trabajos_lugar.append('')

      descripcion= trabajos.find('div', attrs={'class': 'desc'})
      if descripcion:
        trabajos_descripcion.append(descripcion.text)
      else:  trabajos_descripcion.append('')

      hacecuanto= trabajos.find('ul', attrs={'class': 'tags'})
      if hacecuanto:
        trabajos_hacecuanto.append(hacecuanto.text)
      else:  trabajos_hacecuanto.append('')


```

```python
#Importamos pandas para crear el DataFrame asigando un nombre a cada columna con su respectivo lista creada anteriormente
import pandas as pd
# Le asignamos un nombre a cada columna 
trabajos_list = pd.DataFrame({
                               'Entidad': trabajos_cargo,
                               'Compañia': trabajos_compania,
                               'Lugar': trabajos_lugar,
                              'Descripcion': trabajos_descripcion,
                              'Hace_Cuanto': trabajos_hacecuanto
                              })

trabajos_list
# como se puede ver, se obtubieron los datos en el dataframe, sin embargo queda hacer una limpieza de estos 
```

#**Limpieza del Dataframe**

Limpieza de columna Entidad

```python
# Eliminamos "\n"de la columna Entidad con el metodo replace(), reemplanzandolo por una cadena vacia
trabajos_list.Entidad = trabajos_list.Entidad.str.replace('\n','')
trabajos_list
```

Limpieza de columna Lugar

```python
# Eliminamos "\n"de la columna Lugar con el metodo replace(), reemplanzandolo por una cadena vacia
trabajos_list.Lugar = trabajos_list.Lugar.str.replace('\n','')
trabajos_list
```

Limpieza de columna Descripción

```python
# Hacemos el mismo procedimiento en la columna Descripcion
trabajos_list.Descripcion	 = trabajos_list.Descripcion	.str.replace('\n','')
trabajos_list.Descripcion	 = trabajos_list.Descripcion	.str.replace('-','')
trabajos_list
```

Limpieza de columna

```python
# Hacemos el mismo procedimiento en la columna Hace_cuanto
trabajos_list.Hace_Cuanto = trabajos_list.Hace_Cuanto.str.replace('\n','')
trabajos_list
```

```python
# Creamos el dataframe en formato .csv
trabajos_list.to_csv('OfertasEmpleos.csv', index=None, header=True, encoding='utf-8-sig' )
trabajos_list
```
