```python
url_to_srap = 'https://www.convocatoriasdetrabajo.com/buscar-trabajos-en-LA-LIBERTAD-13.html'
from selenium import webdriver
driver.get(url_to_srap)
```

```python
html_code = driver.page_source
```

```python
# Importamos y hacemos uso de la libreria BeautifulSoup
from bs4 import BeautifulSoup
```

```python
# Identificamos la etiqueta y clase que contiene la informacion que deseamos y con la libreria mencionada se define
soup = BeautifulSoup(html_code,'lxml')
jobs =soup.find_all("article", class_="convocatoria")
```

```python
# Declaramos una lista vacia la cual sirve como inicializados para nuestro estructura For la cual ira extrayendo cada dato que encuentre en determinada etiqueta y clase ya mencionada
trabajos = []
for job in jobs:
  trabajos.append(job.text)
driver.quit()
```

```python
# Importamos la libreria Pandas para trabajar con el dataframe
import pandas as pd
# Le asignamos un nombre a cada columna 
Empleo= pd.DataFrame({'trabajo': trabajos})
Empleo
```

**limpieza de datos**

```python
# Aqui comenzamos el proceso de la limpieza de datos extraidos
# comenzamos separando las cadenas de texto a partir de "\n"
prueba = Empleo["trabajo"].str.split('\n', expand=True)
prueba
```

```python
#Seleccionamos las columnas que necesitamos
prueba2 = prueba[[2,5,6,]]
prueba2 
```

```python
# Le ponemos un nombre a las columnas
prueba2 = prueba2.rename(columns={2:'Empleo'})
prueba2 = prueba2.rename(columns={5:'Finaliza'})
prueba2 = prueba2.rename(columns={6:'Grado'})
prueba2
```

```python
# Separamos las cadenas de texto de la columna Empleo a partir del ":"
prueba3 = prueba2["Empleo"].str.split(':', expand=True)
prueba3
```

```python
#Seleccionamos las columnas que necesitamos
prueba4 = prueba3[[0,2]]
prueba4 
```

```python
# Concatenamos las columnas que necesitamos
prueba5 = pd.concat([prueba4,prueba2], axis=1)
prueba5 
```

```python
# le asignamos nombres a las nuevas columnas creadas
prueba6 = prueba5.rename(columns={0:'Entidad'})
prueba6 = prueba6.rename(columns={2:'Cargo'})
prueba6
```

Limpieza de la columna Finaliza

```python
# Limpiamos la columna Finaliza, eliminando el texto que se repite en esta columna, obteniendo solo la fecha de formato DD/MM//AAAA
prueba6.Finaliza = prueba6.Finaliza.str.replace('Finaliza el','')
prueba6
```

```python
# Eliminamos una columna que no necesitamos
Empleo=prueba6.drop(['Empleo'], axis=1)
Empleo
```

```python
# Creamos el dataframe en formato .csv
Empleo.to_csv('trabajosConvocatoria.csv', index=None, header=True, encoding='utf-8-sig')
Empleo
```
